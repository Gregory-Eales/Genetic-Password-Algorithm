```
# import dependencies
import random
import time
```

```
# define variables
gene_set = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz!, "
target = "Hello, World!"
```

```
# define genetic algorithm
class GeneticAlgorithm(object):
  
  # initiate class attributes
  def __init__(self, target, gene_set):
    self.target = target
    self.gene_set = gene_set
    self.parent = None
    
  # generates parent organisms
  def generate_parent(self):
    self.parent = []
    for i in range(len(self.target)):
      self.parent.append(gene_set[random.randrange(0, len(self.gene_set))])
    self.parent = "".join(self.parent)
    return self.parent
  
  # evaluates the fitness of an organism
  def evaluate_fitness(self, organism):
    fitness = 0
    for i in range(len(self.target)):
      if self.target[i] == organism[i]:
        fitness = fitness + 1
    return fitness
  
  # mutates organism by randomly changing a character
  def mutate_organism(self):
    index = random.randrange(0, len(self.target))
    child = list(self.parent)
    child[index] = self.gene_set[random.randrange(0, len(self.gene_set))]
    return "".join(child)
  
  def display(self):
    print(self.parent, self.evaluate_fitness(self.parent), time.time() - t)
  
  # evolves organism for maximum fitness
  def evolve(self):
    t = time.time()
    self.parent = self.generate_parent()
    while True:
      self.child = self.mutate_organism()
      if self.evaluate_fitness(self.parent) == len(self.target):
        self.display()
        break
      if self.evaluate_fitness(self.child) >= self.evaluate_fitness(self.parent):
        self.parent = self.child
        self.display()
      
```

```
GA = GeneticAlgorithm(target, gene_set)
GA.evolve()
```
